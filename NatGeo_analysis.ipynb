```python
import pandas as pd
import re,string
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
nat_geo = pd.read_csv('nat_geo.csv', thousands=",")
labels = pd.read_csv('nat_geo_labes_scores.csv')
```

```python
nat_geo['Likes'] = nat_geo['Likes'].astype(int)
nat_geo['normalized_likes'] = nat_geo['Likes']/max(nat_geo['Likes'])

def remove_k(s):
    """This functions removes the K at the end of numbers and convert them to float. E.g. 12.5K turns to 12500"""
    
    if 'k' in s:
        s=float(s.rstrip('k'))*1000
    else:
        s = float(s.replace(",", ''))
    return s

nat_geo['Comments_clean'] = nat_geo['Comments'].map(remove_k) 
nat_geo['Comments_normalized'] = nat_geo['Comments_clean']/max(nat_geo['Comments_clean'])
```

```python
nat_geo['score'] = 0.4*nat_geo['normalized_likes'] + 0.6*nat_geo['Comments_normalized']
```

```python
nat_geo.head()
```

```python
nat_geo['Image'] = nat_geo['Image'].str.replace(".jpg", '')
```

#### Merging the two datasets

```python
df = pd.merge(nat_geo, labels[['imageid','labels']], left_on= 'Image', right_on= 'imageid')
```

#### Cleaning the Labels

```python
new_labels= df['labels'].map(lambda x: x[1:-1].replace("'","").split(', '))
```

```python
def combine_labels(li):
    """This function combines the tokens in label column, which makes it easy to pass in Countvectorizer"""
    q =''
    for k in li:
        q = q+' '+ k
    return q

df['complete_lables'] = new_labels.map(combine_labels)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(df['complete_lables'])
count_vect.get_feature_names()
X_matrix = X_train_counts.todense()
```

```python
y = np.array(df['score'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_matrix, y, test_size=0.3, random_state=1)
X_matrix.shape
```

```python
rf_reg = RandomForestRegressor(n_estimators=100, max_features=0.333)
rf_reg.fit(X_train, y_train)
```

```python
y_pred_rf = rf_reg.predict(X_test)
rmse_rf = np.sqrt(np.mean((y_pred_rf-y_test)**2))
print("RMSE for 'Only Labels' regression is: ", rmse_rf)
```

```python
import re
from nltk.corpus import stopwords
```

```python
emoji_regexp = u'[\U00002600-\U000026FF]|[\U00002700-\U000027BF]|[\U0001f300-\U0001f5fF]|[\U0001f600-\U0001f64F]|' \
               u'[\U0001f680-\U0001f6FF]|[\U00002A00-\U00002AFF]|[\U0001F1E0-\U0001F1FF]'
```

```python
def cleanup(sentence):
    #letters_only = re.sub("[^a-zA-Z]", " ", sentence.split(', ')[0])
    #emojis = re.findall(emoji_regexp, sentence.split(', ')[0], re.UNICODE)
    try:
        letters_only = re.sub("[^a-zA-Z]", " ", sentence)
        emojis = re.findall(emoji_regexp, sentence, re.UNICODE)
    except:
        letters_only = ''
        emojis = []
    words = letters_only.lower().split()        
    words = [w for w in words if not w in stopwords.words("english")]
    words = [w for w in words if len(w) > 1]
    return (" ".join(words)), (" ".join(emojis))
```

```python
clean_captions = []
clean_emojis = []
for i in range( 0, df['Caption'].size ):
    if( (i+1)%100 == 0 ):
        print(str(round(i/1000*100))+"% done")
    x,y = cleanup(df['Caption'][i])
    clean_captions.append(x)
    clean_emojis.append(y)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
def tokenize(text):
    tokens = nltk.word_tokenize(text)
    stems = []
    for item in tokens:
        stems.append(PorterStemmer().stem(item))
    return stems
```

```python
clean_vectorizer = CountVectorizer(analyzer = "word", tokenizer = tokenize, preprocessor = None, stop_words = 'english') 

# fit_transform() does two functions: First, it fits the model
# and learns the vocabulary; second, it transforms our training data
# into feature vectors. The input to fit_transform should be a list of 
# strings.
clean_features = clean_vectorizer.fit_transform(clean_captions)

#tfidf_transformer = TfidfTransformer()
#clean_features = tfidf_transformer.fit_transform(clean_features)

# Numpy arrays are easy to work with, so convert the result to an 
# array
clean_features = clean_features.toarray()
```

```python
vocab = clean_vectorizer.get_feature_names()
#print(vocab)

new_vocab = []
for word in vocab:
    if len(word)>5:
        new_vocab.append(word)
print (' '.join(new_vocab))
```

```python
clean_features = pd.DataFrame(data=clean_features, columns=vocab)
clean_features = clean_features[new_vocab]
clean_features.shape
```

```python
y = np.array(df['score'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(clean_features, y, test_size=0.3, random_state=1)
rf_reg = RandomForestRegressor(n_estimators=500, max_features=500)

rf_reg.fit(X_train, y_train)

y_pred_rf = rf_reg.predict(X_test)
```

##### Coverting string input to a list

##### Carry out Bag of words on input 

```python
dictionary = gensim.corpora.Dictionary(new_labels)
count = 0
for k, v in dictionary.iteritems():
    print(k, v)
    count += 1
    if count > 10:
        break
```

##### Gensim doc2bow: For each document we create a dictionary reporting how many words and how many times those words appear

```python
bow_corpus = [dictionary.doc2bow(doc) for doc in new_labels]
bow_corpus[20]
```

#### TF-IDF : tf-idf model object created using models.TfidfModel on ‘bow_corpus’ saved to ‘tfidf’. Transformations applied to the entire corpus and called ‘corpus_tfidf’.

```python
from gensim import corpora, models
tfidf = models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]
from pprint import pprint
for doc in corpus_tfidf:
    pprint(doc)
    break
```

##### Running LDA using Bag of Words

```python
lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics=4, id2word=dictionary, passes=2, workers=2)
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
    print('\n')
```

##### Running LDA using TF-IDF

```python
lda_model_tfidf = gensim.models.LdaMulticore(corpus_tfidf, num_topics=4, id2word=dictionary, passes=2, workers=4)
for idx, topic in lda_model_tfidf.print_topics(-1):
    print('Topic: {} Word: {}'.format(idx, topic))
    print('\n')
```

```python
df_topics = pd.DataFrame(columns = ['topic_0','topic_1','topic_2','topic_3'])
for idx,topic in enumerate(lda_model_tfidf.get_document_topics(corpus_tfidf)):
    #print(idx)
    #print(topic[0][1])
    df_topics.loc[len(df_topics)] = [topic[0][1],topic[1][1],topic[2][1],topic[3][1]]
    #df_topics[idx]['topic_1'] = topic[1][1]
    #df_topics[idx]['topic_2']= topic[2][1]
    #df_topics[idx]['topic_3'] = topic[3][1]
```
